```python
import csv
import pandas as pd
import numpy as np

#input_file = 'c:/Users/kmentzer/desktop/electiononly_tweets.csv'
# If restarting process then load this file which has the data up through step 4.
input_file = 'miatwitterprojectCSV.csv'

data = pd.read_csv('miatwitterprojectCSV.csv', error_bad_lines=False)
data = pd.read_csv(input_file, header = 0)
df = pd.DataFrame(data)

print(df.head())
print(df.tail())
```

```python
pd.set_option('display.max_columns', None) # This will make it so all columns appear when printing
pd.options.display.max_colwidth = 300  # This makes it so we can see the entire tweet text.
pd.set_option('display.max_rows', 1000)
print(df.head())
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
lv_sent = []
analyzer = SentimentIntensityAnalyzer()

for index, row in df.iterrows():
    lv_tweet = str(row['text'])
    vs = analyzer.polarity_scores(lv_tweet)
    lv_sent.append(vs)
    
df['VaderSentiment'] = lv_sent
```

```python
df.head()
```

```python
lv_pfz = []
lv_mda = []

for index, row in df.iterrows():
    lv_tweet = str(row['text'].lower())
    
    if "Pfizer" in lv_tweet:
        lv_pfz.append(1)
    else:
        lv_pfz.append(0)

    if "Moderna" in lv_tweet:
        lv_mda.append(1)
    else:
        lv_mda.append(0)

        
df['Pfizer'] = lv_pfz
df['Moderna'] = lv_mda
```

```python
df.head()
```

```python
# Lets extract just the compound score
compound = []
for index, row in df.iterrows():
   # compund.append(row['VaderSentiment']['compound'])
   compound.append(row['VaderSentiment']['compound'])

df['VCompound'] = compound
```

```python
df.groupby('Pfizer')['VCompound'].agg(['count','mean','std'])
```
